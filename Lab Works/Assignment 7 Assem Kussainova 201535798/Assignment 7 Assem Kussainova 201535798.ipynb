```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
import numpy as np
import random
```

```python
image = mpimg.imread('lena.png') 
plt.imshow(image) 
```

# Task 1

```python
def shift_right(img, n):
    img2 = np.zeros(img.shape)
    img2[:, n:] = img[:, :-n]
    plt.imshow(img2)
    plt.savefig('shift_right_image.png')

shift_right(image, 100)
```

# Task 2

```python
def rotation(img, angle, num):
    for k in range(num):
        h, w, d = img.shape
        img2 = np.zeros(img.shape)
    
        for i in range(h):
            for j in range(w):
                img2[i][j] = img[j][h-1-i]
        img = img2
        print("Rotation by ", angle, " degrees: ")
        plt.imshow(img2)
        if k == 0:
            plt.savefig('rotation90.png')
        elif k == 1:
            plt.savefig('rotation180.png')
        else:
            plt.savefig('rotation270.png')
        plt.show()
        angle = angle + 90

rotation(image, 90, 3)
```

# Task 3

```python
def flip(img):
    img2 = np.zeros(img.shape)
    img2 = img[:,::-1,:]
    plt.imshow(img2)
    plt.savefig('flip_image.png')
    
flip(image)
```

# Task 4

```python
def remove(img):
    img2 = img.copy()
    h, w, d = img2.shape
    percent_10 = int(np.sqrt((h*w)*0.1))
    region = (percent_10, percent_10) 
    h_region = random.randint(0, h-region[0]-1)
    w_region = random.randint(0, w-region[0]-1)
    for i in range(h_region, region[0] + h_region):
        for j in range(w_region, region[1] + w_region):
            img2[i][j] = [0, 0, 0]
    plt.imshow(img2)
    plt.savefig('remove_10percent.png')

remove(image)    
```
