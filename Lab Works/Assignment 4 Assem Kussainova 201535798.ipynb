```python
import numpy as np
```

```python
def MatrixGenerator():
    arr = np.random.uniform(-10, 10, (100, 9))
    targets = np.random.randint(0, 9, 100)

    arr = arr.tolist()
    targets = targets.tolist()

    for i in range(len(arr)):
        arr[i].append(targets[i])
    return arr, targets
```

# SVM Loss

```python
def LossCount(a):
    l = []
    for i in range(len(a)):
        y = a[i][9]
        maxi = []
        
        for j in range(len(a[i])-1):
            const = a[i][y]
            if j != y:
                maxi.append(np.maximum(0, a[i][j] - const + 1))
                
        loss = sum(maxi)
        l.append(loss)
    return l
```

```python
def SVM_loss(l):
    return np.mean(l)
```

# Cross Entropy Loss

```python
def normalize(m):
    for i in range(len(m)):
        m[i] = m[i][:len(m[i])-1]
        norm = np.zeros(len(m[i]))
        for j in range(len(m[i])):
            norm[j] = m[i][j] - min(m[i]) + 1
        n = sum(norm)
        for j in range(len(norm)):
            if min(norm) != max(norm):
                norm[j] = norm[j]/n            
            else:
                norm[j] = 0.5
        m[i] = norm
    return m
```

```python
def CrossEntropy(m):
    l = []
    
    for i in range(len(m)):
        loss = 0
        j = m[i][-1]
        loss = loss - np.log(m[i][j])
        l.append(loss)
    return np.mean(l)
```

```python
points = []
points2 = []

for i in range(1, 1001):
    matrix, target = MatrixGenerator()
    losses = LossCount(matrix)
    points.append(SVM_loss(losses))
    
    norm_matrix = normalize(matrix)
    
    for i in range(len(norm_matrix)):
        norm_matrix[i] = norm_matrix[i].tolist()
        norm_matrix[i].append(target[i])
            
    points2.append(CrossEntropy(norm_matrix))
```

# Plot

```python
import matplotlib.pyplot as plt

plot1 = plt.figure(1, figsize = (20, 10))
plt.plot(list(range(1, 1001)), points)
plt.plot(list(range(1, 1001)), points2)
plt.title("SVM and Cross Entropy Loss")
plt.xlabel('Iteration')
plt.ylabel('Loss')

plt.show()
```
