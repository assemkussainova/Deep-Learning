```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
my_image = np.genfromtxt('in.txt', delimiter=',')
```

```python
plt.imshow(my_image,cmap='gray')
plt.show()
```

```python
hed_kernel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
hed_kernel
```

```python
ved_kernel = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
ved_kernel
```

```python
def convolution(image, kernel):

    output = np.zeros_like(image)

    image_padded = np.zeros((image.shape[0] + 2, image.shape[1] + 2))
    image_padded[1:-1, 1:-1] = image

    for x in range(image.shape[1]):
        for y in range(image.shape[0]):
            output[y, x] = (kernel * image_padded[y: y+3, x: x+3]).sum()

    return output
```

```python
img1 = my_image.copy()
Final_Result_1 = convolution(img1, hed_kernel)
plt.imshow(Final_Result_1, cmap='gray')
plt.show()
```

```python
img2 = my_image.copy()
Final_Result_2 = convolution(img2, ved_kernel)
plt.imshow(Final_Result_2, cmap='gray')
plt.show()
```
